```python
import math
elementList = ['', 'Si','Ti','Al','Fe','Mg','Mn','Ca','Na','K','O','H','C','Cl','e','Ni','Zr']
f = open('hp-abbrv.txt', 'r')
phaseNames = {}
while True:
    line = f.readline()
    if not line: break
    line = line.split(',')
    phaseNames[line[1].strip('\n')]= line[0].title()
print (phaseNames)
```

```python
outputHeaders = False
f = open('tc-ds55.txt', 'r')
while True:
    line = f.readline()
    if not line: break
    lineOne = ' '.join(line.split()).split(" ")
    lineTwo = ' '.join(f.readline().split()).split(" ")
    lineOneLength = len(lineOne)
    lineTwoLength = len(lineTwo)
    
    name = lineOne[0]
    
    i = 2
    formula = ''
    while lineOne[i] != '0':
        index = int(float(lineOne[i]))
        symbol = elementList[index]
        coeff = float(lineOne[i+1])
        formula += symbol
        if math.ceil(coeff) == math.floor(coeff):
            if int(coeff) > 1:
                formula += str(int(coeff))
        else:
            formula += str(coeff)
        i += 2
    
    h = float(lineOne[lineOneLength-3])*1000.0
    s = float(lineOne[lineOneLength-2])*1000.0
    v = float(lineOne[lineOneLength-1])
    
    a = float(lineTwo[0])*1000.0
    b = float(lineTwo[1])*1000.0
    c = float(lineTwo[2])*1000.0
    d = float(lineTwo[3])*1000.0
    
    a0 = float(lineTwo[4])
    K = float(lineTwo[6])*1000.0
    
    Tc = 0.0
    Smax = 0.0
    Vmax = 0.0
    if (lineTwoLength > 10):
        Tc = float(lineTwo[10])
        Smax = float(lineTwo[11])*1000.0
        Vmax = float(lineTwo[12])
    
    if outputHeaders:
        print ("@interface " + phaseNames[name] + "HollandAndPowell" + ": HollandAndPowellProperties { }")
        print ("@end")
        print ()
    else:
        print ("@implementation " + phaseNames[name] + "HollandAndPowell")
        print ()
        print ("-(id)init {")
        print ("    if ((self = [super initWithH:" + str(h))
        print ("                               S:" + str(s))
        print ("                               a:" + str(a))
        print ("                               b:" + str(b))
        print ("                               c:" + str(c))
        print ("                               d:" + str(d))
        print ("                             Tc0:" + str(Tc))
        print ("                            Smax:" + str(Smax))
        print ("                            Vmax:" + str(Vmax))
        print ("                              v0:" + str(v))
        print ("                              a0:" + str(a0))
        print ("                               K:" + str(K) + "])) {")
        print ("    [self setPhaseFormula:@" + '"' + formula + '"' + "];")
        print ("    [self setPhaseName:@" + '"' + phaseNames[name] + '"' + "];")
        print ("    }")
        print ("    return self;")
        print ("}")
        print ()
        print ("@end")
        print ()
```
