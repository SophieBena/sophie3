### Stixrude opx
Input compositional variables:  
$X_{En}$, $X_{Fs}$, $X_{MaTs}$, $X_{Di}$  

Site mole fractions:  

**(1)** $X_{Ca}^{M2} = X_{Di}$  

$X_{Mg}^{M2} + X_{Mg}^{M1} = 2X_{En} + X_{MaTs} + X_{Di}$  

$X_{Fe}^{M2} + X_{Fe}^{M1} = 2X_{Fs}$  

**(2)** $X_{Al}^{M1} = X_{MaTs}$  

To which we must add a constraint that there is no preferential ordering of Mg and Fe on sites:  

$\frac{X_{Mg}^{M1}}{X_{Fe}^{M1}} = \frac{X_{Mg}^{M2}}{X_{Fe}^{M2}}$  

We recognize that site mole fractions must add to unity:  

$X_{Ca}^{M2} + X_{Mg}^{M2} + X_{Fe}^{M2} = 1$ and  

$X_{Mg}^{M1} + X_{Fe}^{M1} + X_{Al}^{M1} = 1$  

and substituting the uniquely defined site mole fractions  

$X_{Mg}^{M2} + X_{Fe}^{M2} = 1 - X_{Ca}^{M2} = 1 - X_{Di}$ and  

$X_{Mg}^{M1} + X_{Fe}^{M1} = 1 - X_{Al}^{M1} = 1 - X_{MaTs}$  

Next divide the first equation by $X_{Fe}^{M2}$ and the second by $X_{Fe}^{M1}$ to give  

$\frac{X_{Mg}^{M2}}{X_{Fe}^{M2}} + 1 = \frac{1 - X_{Di}}{X_{Fe}^{M2}}$ and  

$\frac{X_{Mg}^{M1}}{X_{Fe}^{M1}} + 1 = \frac{1 - X_{MaTs}}{X_{Fe}^{M1}}$  

The two equations must be equal because of the assumed random ordering condition, hence  

$\frac{1 - X_{Di}}{X_{Fe}^{M2}} = \frac{1 - X_{MaTs}}{X_{Fe}^{M1}}$  or $X_{Fe}^{M1}(1 - X_{Di}) = X_{Fe}^{M2} (1 - X_{MaTs})$  

substituting, $X_{Fe}^{M2} = 2X_{Fs} - X_{Fe}^{M1}$  

$X_{Fe}^{M1}(1 - X_{Di}) = (2X_{Fs} - X_{Fe}^{M1})(1 - X_{MaTs})$   

we have

**(3)** $X_{Fe}^{M1} = 2X_{Fs} \left( \frac{1 - X_{MaTs}}{2 - X_{Di} - X_{MaTs}} \right) = 2X_{Fs} \left( \frac{X_{En} + X_{Fs} + X_{Di}}{2X_{En} + 2X_{Fs} + X_{Di} + X_{MaTs}} \right)$   

and therefore

**(4)** $X_{Fe}^{M2} = 2X_{Fs} \left( 1 - \frac{1 - X_{MaTs}}{2 - X_{Di} - X_{MaTs}} \right) = 2X_{Fs} \left( 1 - \frac{X_{En} + X_{Fs} + X_{Di}}{2X_{En} + 2X_{Fs} + X_{Di} + X_{MaTs}} \right) = 2X_{Fs} \left(\frac{X_{En} + X_{Fs} + X_{MaTs}}{2X_{En} + 2X_{Fs} + X_{Di} + X_{MaTs}} \right)$  

**(5)** $X_{Mg}^{M2}  = 1 - X_{Di} - 2X_{Fs}(1 - \frac{1 - X_{MaTs}}{2 - X_{Di} - X_{MaTs}}) = 1 - X_{Di} - 2X_{Fs} \left(\frac{X_{En} + X_{Fs} + X_{MaTs}}{2X_{En} + 2X_{Fs} + X_{Di} + X_{MaTs}} \right)$  and  

**(6)** $X_{Mg}^{M1} = 1 - X_{MaTs} - 2X_{Fs} \frac{1 - X_{MaTs}}{2 - X_{Di} - X_{MaTs}} = 1 - X_{MaTs} - 2X_{Fs} \left( \frac{X_{En} + X_{Fs} + X_{Di}}{2X_{En} + 2X_{Fs} + X_{Di} + X_{MaTs}} \right)$  

### Stixrude cpx
Input compositional variables:  
$X_{Di}$, $X_{Hd}$, $X_{En}$, $X_{CaTs}$, $X_{Jd}$  

Site mole fractions:  

**(1)** $X_{Ca}^{M2} = X_{Di} + X_{Hd} + X_{CaTs}$  

$X_{Mg}^{M2} + X_{Mg}^{M1} = X_{Di} + 2X_{En}$  

$X_{Fe}^{M2} + X_{Fe}^{M1} = X_{Hd}$  

**(2)** $X_{Al}^{M1} = X_{CaTs} + X_{Jd}$  

**(3)** $X_{Na}^{M2} = X_{Jd}$  

To which we must add a constraint that there is no preferential ordering of Mg and Fe on sites:  

$\frac{X_{Mg}^{M1}}{X_{Fe}^{M1}} = \frac{X_{Mg}^{M2}}{X_{Fe}^{M2}}$  

We recognize that site mole fractions must add to unity:  

$X_{Na}^{M2} + X_{Ca}^{M2} + X_{Mg}^{M2} + X_{Fe}^{M2} = 1$ and  

$X_{Mg}^{M1} + X_{Fe}^{M1} + X_{Al}^{M1} = 1$  

substituting  

$X_{Mg}^{M1} + X_{Fe}^{M1} = 1 - X_{Al}^{M1} = 1 - X_{CaTs} - X_{Jd} = A$  

$X_{Mg}^{M2} + X_{Fe}^{M2} = 1 - X_{Na}^{M2} - X_{Ca}^{M2} = 1 - X_{Jd} - X_{Di} - X_{Hd} - X_{CaTs} = B$  

then rearranging and inserting $X_{Fe}^{M2} + X_{Fe}^{M1} = X_{Hd}$,  

**(4)** $X_{Fe}^{M2} = X_{Hd} \frac{B}{A+B}  = X_{Hd} \frac{1 - X_{Jd} - X_{Di} - X_{Hd} - X_{CaTs}}{2 - 2X_{CaTs} - 2X_{Jd} - X_{Di} - X_{Hd}} = X_{Hd} \frac{X_{En}}{X_{Di} + X_{Hd} + 2X_{En}}$  

**(5)** $X_{Fe}^{M1} = X_{Hd} \left(1 - \frac{1 - X_{Jd} - X_{Di} - X_{Hd} - X_{CaTs}}{2 - 2X_{CaTs} - 2X_{Jd} - X_{Di} - X_{Hd}} \right) = X_{Hd} \left(1 - \frac{X_{En}}{X_{Di} + X_{Hd} + 2X_{En}} \right) = X_{Hd} \left(\frac{X_{Di} + X_{Hd} + X_{En}}{X_{Di} + X_{Hd} + 2X_{En}} \right)$  

**(6)** $X_{Mg}^{M1} = 1 - X_{Al}^{M1} = 1 - X_{CaTs} - X_{Jd} - X_{Hd} \left(1 - \frac{1 - X_{Jd} - X_{Di} - X_{Hd} - X_{CaTs}}{2 - 2X_{CaTs} - 2X_{Jd} - X_{Di} - X_{Hd}} \right) = X_{Di} + X_{Hd} + X_{En} - X_{Hd} \left(\frac{X_{Di} + X_{Hd} + X_{En}}{X_{Di} + X_{Hd} + 2X_{En}} \right)$  

**(7)** $X_{Mg}^{M2} = 1 - X_{Na}^{M2} - X_{Ca}^{M2} = 1 - X_{Jd} - X_{Di} - X_{Hd} - X_{CaTs} - X_{Hd} \frac{1 - X_{Jd} - X_{Di} - X_{Hd} - X_{CaTs}}{2 - 2X_{CaTs} - 2X_{Jd} - X_{Di} - X_{Hd}} = X_{En} - X_{Hd} \frac{X_{En}}{X_{Di} + X_{Hd} + 2X_{En}}$  



### Stixrude garnet

Input compositional variables  

$X_{Py}$, $X_{Al}$, $X_{Gr}$, $X_{Mj}$, $X_{Na}$  

$3 X_{Mg}^{dod} + 2 X_{Mg}^{oct} = 3 X_{Py} + 4 X_{Mj}$  

$3 X_{Fe}^{dod} + 2 X_{Fe}^{oct} = 3 X_{Al}$  

**(1)** $3 X_{Ca}^{dod} = 3 X_{Gr}$  

**(2)** $3 X_{Na}^{dod} = 2 X_{Na}$  

**(3)** $3 X_{Al}^{dod} = X_{Na}$  

**(4)** $2 X_{Al}^{oct} = 2 X_{Py} + 2 X_{Al} + 2 X_{Gr} + X_{Na}$  

**(5)** $2 X_{Si}^{oct} = X_{Mj} + X_{Na}$  

site balance  

$X_{Mg}^{dod} + X_{Fe}^{dod} + X_{Ca}^{dod} + X_{Na}^{dod} + X_{Al}^{dod} = 1$  

$X_{Mg}^{oct} + X_{Fe}^{oct} + X_{Al}^{oct} + X_{Si}^{oct} = 1$  

hence  

$X_{Mg}^{dod} + X_{Fe}^{dod} = 1 -  X_{Ca}^{dod} - X_{Na}^{dod} - X_{Al}^{dod} = 1 - X_{Gr} - \frac{2}{3} X_{Na} - \frac{1}{3} X_{Na} = 1 -  X_{Gr} - X_{Na}$  

$X_{Mg}^{oct} + X_{Fe}^{oct} = 1 - X_{Al}^{oct} - X_{Si}^{oct} = 1 - X_{Py} - X_{Al} - X_{Gr} - \frac{1}{2} X_{Na} - \frac{1}{2} X_{Mj} - \frac{1}{2} X_{Na} = \frac{1}{2} X_{Mj}$  

we assume no Fe-Mg ordering  

$\frac{X_{Mg}^{dod}}{X_{Fe}^{dod}} = \frac{X_{Mg}^{oct}}{X_{Fe}^{oct}}$  

from which  

$X_{Fe}^{oct} \left( 1 -  X_{Gr} - X_{Na} \right) = X_{Fe}^{dod} \frac{1}{2} X_{Mj}$  

since $X_{Fe}^{dod} = X_{Al} - \frac{2}{3} X_{Fe}^{oct}$  

$X_{Fe}^{oct} \left( 1 -  X_{Gr} - X_{Na} + \frac{1}{3} X_{Mj} \right) = \frac{1}{2} X_{Al} X_{Mj}$  

so

**(6)** $X_{Fe}^{oct} = \frac{1}{2} \frac{X_{Al} X_{Mj}}{1 -  X_{Gr} - X_{Na} + \frac{1}{3} X_{Mj}}$  

**(7)** $X_{Fe}^{dod} = X_{Al} \frac{1 -  X_{Gr} - X_{Na}}{1 - X_{Gr} - X_{Na} + \frac{1}{3} X_{Mj}}$  

**(8)** $X_{Mg}^{oct} = \frac{1}{2} X_{Mj} - \frac{1}{2} \frac{X_{Al} X_{Mj}}{1 -  X_{Gr} - X_{Na} + \frac{1}{3} X_{Mj}} = \frac{1}{2} X_{Mj} \left( 1 - \frac{X_{Al}}{1 -  X_{Gr} - X_{Na} + \frac{1}{3} X_{Mj}} \right)$  

**(9)** $X_{Mg}^{dod} = 1 -  X_{Gr} - X_{Na} - X_{Al} \frac{1 -  X_{Gr} - X_{Na}}{1 - X_{Gr} - X_{Na} + \frac{1}{3} X_{Mj}} = \left( 1 -  X_{Gr} - X_{Na} \right) \left( 1 - \frac{X_{Al}}{1 - X_{Gr} - X_{Na} + \frac{1}{3} X_{Mj}} \right)$
