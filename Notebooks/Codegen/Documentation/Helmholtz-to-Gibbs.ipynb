# Derivative relations for a Helmholtz Model
This notebook provides a summary of conversion relations between derivatives of the Helmholtz free energy and teh Gibbs free energy, each expressed in terms of natural variables (*T*, *V*) and (*T*, *P*), respectively.

### Autogenerated Helmholtz free energy derivatives:
${\left( {\frac{{\partial A}}{{\partial T}}} \right)_V}$
${\left( {\frac{{\partial A}}{{\partial V}}} \right)_T}$
${\left( {\frac{{{\partial ^2}A}}{{\partial {T^2}}}} \right)_V}$
$\left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right)$
${\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)_T}$
${\left( {\frac{{{\partial ^3}A}}{{\partial {T^3}}}} \right)_V}$
$\left( {\frac{{{\partial ^3}A}}{{\partial {T^2}\partial V}}} \right)$
$\left( {\frac{{{\partial ^3}A}}{{\partial T\partial {V^2}}}} \right)$
${\left( {\frac{{{\partial ^3}A}}{{\partial {V^3}}}} \right)_T}$

### Identities that relate the Helmholtz and Gibbs free energies:
$dA =  - SdT - PdV = {\left( {\frac{{\partial A}}{{\partial T}}} \right)_V}dT + {\left( {\frac{{\partial A}}{{\partial V}}} \right)_T}dV$  
$G = A + PV$  
$dG = dA + PdV + VdP$  
$dG =  - SdT + VdP = {\left( {\frac{{\partial A}}{{\partial T}}} \right)_V}dT + VdP$

### We seek definitions for the following derivatives of the Gibbs free energy:

${\left( {\frac{{\partial G}}{{\partial T}}} \right)_P}$
${\left( {\frac{{\partial G}}{{\partial P}}} \right)_T}$
${\left( {\frac{{{\partial ^2}G}}{{\partial {T^2}}}} \right)_P}$
$\left( {\frac{{{\partial ^2}G}}{{\partial T\partial P}}} \right)$
${\left( {\frac{{{\partial ^2}G}}{{\partial {P^2}}}} \right)_T}$
${\left( {\frac{{{\partial ^3}G}}{{\partial {T^3}}}} \right)_P}$
$\left( {\frac{{{\partial ^3}G}}{{\partial {T^2}\partial P}}} \right)$
$\left( {\frac{{{\partial ^3}G}}{{\partial T\partial {P^2}}}} \right)$
${\left( {\frac{{{\partial ^3}G}}{{\partial {P^3}}}} \right)_T}$

### (1) ${\left( {\frac{{\partial G}}{{\partial T}}} \right)_P}$
${\left( {\frac{{\partial G}}{{\partial T}}} \right)_P} = {\left( {\frac{{\partial A}}{{\partial T}}} \right)_V} =  - S$

### (2) ${\left( {\frac{{\partial G}}{{\partial P}}} \right)_T}$
${\left( {\frac{{\partial G}}{{\partial P}}} \right)_P} = V$

#### Some definitions:  
$P =  - {\left( {\frac{{\partial A}}{{\partial V}}} \right)_T}$  
$dP =  - d{\left( {\frac{{\partial A}}{{\partial V}}} \right)_T} =  - \left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right)dT - {\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)_T}dV$  
${\left( {\frac{{\partial P}}{{\partial V}}} \right)_T} =  - {\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)_T}$ and ${\left( {\frac{{{\partial ^2}G}}{{\partial {P^2}}}} \right)_T} = {\left( {\frac{{\partial V}}{{\partial P}}} \right)_T}$  
$0 =  - \left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right) - {\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)_T}{\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}$  
${\left( {\frac{{\partial V}}{{\partial T}}} \right)_P} =  - \frac{{\left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right)}}{{{{\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)}_T}}}$ and $\left( {\frac{{{\partial ^2}G}}{{\partial T\partial P}}} \right) = {\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}$

#### From:  
$d{\left( {\frac{{\partial G}}{{\partial T}}} \right)_P} = d{\left( {\frac{{\partial A}}{{\partial T}}} \right)_V} = {\left( {\frac{{{\partial ^2}A}}{{\partial {T^2}}}} \right)_V}dT + \left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right)dV$ and  
${\left( {\frac{{{\partial ^2}G}}{{\partial {T^2}}}} \right)_P} = {\left( {\frac{{{\partial ^2}A}}{{\partial {T^2}}}} \right)_V} + \left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right){\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}$
### (3) ${\left( {\frac{{{\partial ^2}G}}{{\partial {T^2}}}} \right)_P}$
${\left( {\frac{{{\partial ^2}G}}{{\partial {T^2}}}} \right)_P} = {\left( {\frac{{{\partial ^2}A}}{{\partial {T^2}}}} \right)_V} - \frac{{{{\left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right)}^2}}}{{{{\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)}_T}}}$  
#### Note:
${\left( {\frac{{{\partial ^2}A}}{{\partial {T^2}}}} \right)_V} =  - {\left( {\frac{{\partial S}}{{\partial T}}} \right)_V} =  - \frac{{{C_V}}}{T}$ and ${\left( {\frac{{{\partial ^2}G}}{{\partial {T^2}}}} \right)_P} =  - {\left( {\frac{{\partial S}}{{\partial T}}} \right)_P} =  - \frac{{{C_P}}}{T}$

### (4) $\left( {\frac{{{\partial ^2}G}}{{\partial T\partial P}}} \right)$
$\left( {\frac{{{\partial ^2}G}}{{\partial T\partial P}}} \right) =  - \frac{{\left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right)}}{{{{\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)}_T}}}$

### (5) ${\left( {\frac{{{\partial ^2}G}}{{\partial {P^2}}}} \right)_T}$
${\left( {\frac{{{\partial ^2}G}}{{\partial {P^2}}}} \right)_T} = {\left( {\frac{{\partial V}}{{\partial P}}} \right)_T} =  - \frac{1}{{{{\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)}_T}}}$

#### Interlude ...   
$d{\left( {\frac{{{\partial ^2}G}}{{\partial {T^2}}}} \right)_P} = d{\left( {\frac{{{\partial ^2}A}}{{\partial {T^2}}}} \right)_V} + {\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}d\left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right) + \left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right)d{\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}$  
$d{\left( {\frac{{{\partial ^2}G}}{{\partial {T^2}}}} \right)_P} = {\left( {\frac{{{\partial ^3}A}}{{\partial {T^3}}}} \right)_V}dT + \left( {\frac{{{\partial ^3}A}}{{\partial {T^2}\partial V}}} \right)dV + {\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}\left[ {\left( {\frac{{{\partial ^3}A}}{{\partial {T^2}\partial V}}} \right)dT + \left( {\frac{{{\partial ^3}A}}{{\partial T\partial {V^2}}}} \right)dV} \right] + \left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right)\left[ {{{\left( {\frac{{{\partial ^2}V}}{{\partial {T^2}}}} \right)}_P}dT + \left( {\frac{{{\partial ^2}V}}{{\partial T\partial P}}} \right)dP} \right]$ or 
### (6) ${\left( {\frac{{{\partial ^3}G}}{{\partial {T^3}}}} \right)_P}$
${\left( {\frac{{{\partial ^3}G}}{{\partial {T^3}}}} \right)_P} = {\left( {\frac{{{\partial ^3}A}}{{\partial {T^3}}}} \right)_V} + 2\left( {\frac{{{\partial ^3}A}}{{\partial {T^2}\partial V}}} \right){\left( {\frac{{\partial V}}{{\partial T}}} \right)_P} + \left( {\frac{{{\partial ^3}A}}{{\partial T\partial {V^2}}}} \right){\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}{\left( {\frac{{\partial V}}{{\partial T}}} \right)_P} + \left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right){\left( {\frac{{{\partial ^2}V}}{{\partial {T^2}}}} \right)_P}$  
#### Note:
${\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}$ is defined above in terms of Helmoltz derivatives.  For the second derivative, ${\left( {\frac{{{\partial ^2}V}}{{\partial {T^2}}}} \right)_P}$, the identity is:   
$0 =  - \left( {\frac{{{\partial ^3}A}}{{\partial {T^2}\partial V}}} \right) - 2\left( {\frac{{{\partial ^3}A}}{{\partial T\partial {V^2}}}} \right){\left( {\frac{{\partial V}}{{\partial T}}} \right)_P} - {\left( {\frac{{{\partial ^3}A}}{{\partial {V^3}}}} \right)_T}{\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}{\left( {\frac{{\partial V}}{{\partial T}}} \right)_P} - {\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)_T}{\left( {\frac{{{\partial ^2}V}}{{\partial {T^2}}}} \right)_P}$  

For reference, derivatives of the heat capacity are obtained as:  
${\left( {\frac{{{\partial ^3}G}}{{\partial {T^3}}}} \right)_P} =  - {\left( {\frac{{\partial S}}{{\partial T}}} \right)_P} = \frac{1}{T}\left[ {\frac{{{C_P}}}{T} - \left( {\frac{{\partial {C_P}}}{{\partial T}}} \right)} \right]$ and ${\left( {\frac{{{\partial ^3}A}}{{\partial {T^3}}}} \right)_V} = \frac{1}{T}\left[ {\frac{{{C_V}}}{T} - {{\left( {\frac{{\partial {C_V}}}{{\partial T}}} \right)}_V}} \right]$
### (7) $\left( {\frac{{{\partial ^3}G}}{{\partial {T^2}\partial P}}} \right)$
$\left( {\frac{{{\partial ^2}G}}{{\partial {T^2}\partial P}}} \right) = \left( {\frac{{{\partial ^3}A}}{{\partial {T^2}\partial V}}} \right){\left( {\frac{{\partial V}}{{\partial P}}} \right)_T} + \left( {\frac{{{\partial ^3}A}}{{\partial T\partial {V^2}}}} \right){\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}{\left( {\frac{{\partial V}}{{\partial P}}} \right)_T} + \left( {\frac{{{\partial ^2}A}}{{\partial T\partial V}}} \right)\left( {\frac{{{\partial ^2}V}}{{\partial T\partial P}}} \right)$  
#### Note:
${\left( {\frac{{\partial V}}{{\partial P}}} \right)_T}$ is defined above in terms of Helmoltz derivatives.  For the second derivative, $\left( {\frac{{{\partial ^2}V}}{{\partial T\partial P}}} \right)$, the identity is:  
$0 =  - \left( {\frac{{{\partial ^3}A}}{{\partial T\partial {V^2}}}} \right){\left( {\frac{{\partial V}}{{\partial P}}} \right)_T} - {\left( {\frac{{{\partial ^3}A}}{{\partial {V^3}}}} \right)_T}{\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}{\left( {\frac{{\partial V}}{{\partial P}}} \right)_T} - {\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)_T}\left( {\frac{{{\partial ^2}V}}{{\partial T\partial P}}} \right)$
### (8) $\left( {\frac{{{\partial ^3}G}}{{\partial T\partial {P^2}}}} \right)$
$\left( {\frac{{{\partial ^3}G}}{{\partial T\partial {P^2}}}} \right) = \left( {\frac{{{\partial ^2}V}}{{\partial T\partial P}}} \right) = \frac{1}{{{{\left[ {{{\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)}_T}} \right]}^2}}}\left[ {\left( {\frac{{{\partial ^3}A}}{{\partial T\partial {V^2}}}} \right) + {{\left( {\frac{{{\partial ^3}A}}{{\partial {V^3}}}} \right)}_T}{{\left( {\frac{{\partial V}}{{\partial T}}} \right)}_P}} \right]$  
#### Note:
${\left( {\frac{{\partial V}}{{\partial T}}} \right)_P}$ is defined above in terms of Helmoltz derivatives.
### (9) ${\left( {\frac{{{\partial ^3}G}}{{\partial {P^3}}}} \right)_T}$  
${\left( {\frac{{{\partial ^3}G}}{{\partial {P^3}}}} \right)_T} = {\left( {\frac{{{\partial ^2}V}}{{\partial {P^2}}}} \right)_T} = \frac{{{{\left( {\frac{{\partial V}}{{\partial P}}} \right)}_T}}}{{{{\left( {\frac{{{\partial ^2}A}}{{\partial {V^2}}}} \right)}_T}}}$  
#### Note:
${\left( {\frac{{\partial V}}{{\partial P}}} \right)_T}$ is defined above in terms of Helmoltz derivatives.
